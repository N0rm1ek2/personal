```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
#Filter the DataFrame
only_movies = netflix_df[netflix_df['type']=='Movie']
less_than_two_k = only_movies[only_movies['release_year']<2000]
movies_in_the_nineties = less_than_two_k[less_than_two_k['release_year']>=1990]

#Plot the chart with histogram
plt.hist(movies_in_the_nineties['duration'])
plt.title('Distribution of movies duration in the 90s')
plt.xlabel('Duration')
plt.ylabel('# of Movies')
plt.show()
```

```python
#Estimation of most popular duration
duration = 100
```

```python
#Filter the Action movies
action_movies = movies_in_the_nineties[movies_in_the_nineties['genre'] == 'Action']

#Count the short Action movies
short_movie_count = 0

for dur in action_movies['duration']:
    if dur < 90:
        short_movie_count += 1
    else:
        short_movie_count += 0
        
print(short_movie_count)
```
